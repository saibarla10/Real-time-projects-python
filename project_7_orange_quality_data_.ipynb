```python
# 1) Is there a correlation between the size (Size (cm)) and the weight (Weight (g)) of the fruit?
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(data[['Size (cm)','Weight (g)']].corr(), annot=True, cmap='coolwarm')
plt.title('correlation between the size (Size (cm)) and the weight (Weight (g))')
plt.show()
```

```python
# 2) How does the sweetness (Brix) of the fruit correlate with its acidity (pH)? Do sweeter fruits have lower acidity?
```

```python
brix = data['Brix (Sweetness)']
ph = data['pH (Acidity)']
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(brix, ph, alpha=0.5, color='blue')
plt.title('Sweetness (Brix) vs Acidity (pH)')
plt.xlabel('Brix (Sweetness)')
plt.ylabel('pH (Acidity)')
```

```python
# 3) How does the softness (Softness (1-5)) of the fruit relate to its ripeness (Ripeness (1-5))?
```

```python
softness = data['Softness (1-5)']
ripeness = data['Ripeness (1-5)']
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(softness, ripeness, alpha=0.5)
plt.title('Softness vs Ripeness')
plt.xlabel('Softness (1-5)')
plt.ylabel('Ripeness (1-5)')
plt.show()
```

```python
# 4) Do fruits with blemishes (Blemishes (Y/N)) tend to have lower ripeness or softness levels?
```

```python
data['Blemishes (Y/N)'] = data['Blemishes (Y/N)'].apply(lambda x: 'Y' if 'Y' in x else 'N')
```

```python
blemished = data[data['Blemishes (Y/N)'] == 'Y']
no_blemishes = data[data['Blemishes (Y/N)'] == 'N']
```

```python
blemished_ripeness_mean = blemished['Ripeness (1-5)'].mean()
no_blemishes_ripeness_mean = no_blemishes['Ripeness (1-5)'].mean()
blemished_softness_mean = blemished['Softness (1-5)'].mean()
no_blemishes_softness_mean = no_blemishes['Softness (1-5)'].mean()
```

```python
print("Average Ripeness (1-5):")
print(f"With Blemishes: {blemished_ripeness_mean:.2f}")
print(f"Without Blemishes: {no_blemishes_ripeness_mean:.2f}")
print("\nAverage Softness (1-5):")
print(f"With Blemishes: {blemished_softness_mean:.2f}")
print(f"Without Blemishes: {no_blemishes_softness_mean:.2f}")
```
