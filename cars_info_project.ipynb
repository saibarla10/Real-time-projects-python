```python
# Line plot
```

```python
avg_price_by_year = data.groupby('Year')['Price'].mean().reset_index()

sns.lineplot(data=avg_price_by_year, x='Year', y='Price', marker='o')
plt.title('Average Car Price by Year')
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Bar chart or plot
```

```python
avg_price_by_fuel = data.groupby('Fuel')['Price'].mean().reset_index()

sns.barplot(data=avg_price_by_fuel, x='Fuel', y='Price')
plt.title('Average Car Price by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# scatter plot
```

```python
sns.scatterplot(data=data, x='Kilometers', y='Price')
plt.title('Price vs Kilometers Driven')
plt.xlabel('Kilometers Driven')
plt.ylabel('Price')
plt.tight_layout()
plt.show()
```

```python
# Histplot
```

```python
sns.histplot(data=data, x='Price', bins=30, kde=True)
plt.title('Distribution of Car Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Boxplot
```

```python
sns.boxplot(data=data, x='Fuel', y='Price')
plt.title('Car Price Distribution by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# violinplot
```

```python
sns.violinplot(data=data, x='Fuel', y='Price', inner='box', scale='width')
plt.title('Car Price Distribution by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# countplot
```

```python
sns.countplot(data=data, x='Fuel')
plt.title('Number of Cars by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# KDE
```

```python
sns.kdeplot(data['Price'])
plt.title("This is KDE")
plt.show()
```

```python
# 1) What is the average price of cars across different brands, and which brands have the highest and lowest average prices?
```

```python
top_brands = data.groupby('Brand')['Price'].mean().sort_values(ascending=False).head(10).reset_index()

sns.barplot(data=top_brands, x='Price', y='Brand')
```

```python
# 2) How does the condition of a car (Status) affect its price?
```

```python
data.groupby('Status')['Price'].describe()
```

```python
sns.boxplot(data=data, x='Status', y='Price')
plt.title('Car Price Distribution by Status')
```

```python
# 3) Is there a correlation between mileage (Kilometers) and the price of a car?
```

```python
sns.scatterplot(data=data, x='Kilometers', y='Price', alpha=0.6)
plt.title('Car Price vs. Kilometers Driven')
plt.xlabel('Kilometers Driven')
plt.ylabel('Price')
plt.tight_layout()
plt.show()
```

```python
sns.regplot(data=data, x='Kilometers', y='Price', scatter_kws={'alpha':0.4})
plt.title('Regression: Price vs. Kilometers')
plt.xlabel('Kilometers Driven')
plt.ylabel('Price')
plt.tight_layout()
plt.show()
```

```python
# 4) How does engine capacity (CC) relate to car price and mileage?
```

```python
sns.scatterplot(data=data, x='CC', y='Price', alpha=0.6)
plt.title('Car Price vs. Engine Capacity')
```

```python
sns.scatterplot(data=data, x='CC', y='Kilometers', alpha=0.6)
plt.title('Mileage vs. Engine Capacity')
```
