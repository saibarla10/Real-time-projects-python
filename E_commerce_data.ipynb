```python
avg_purchase = data.groupby('Product Category')['Total Purchase Amount'].mean().reset_index()

sns.lineplot(data=avg_purchase, x='Product Category', y='Total Purchase Amount', marker='o')
plt.title('Average Purchase Amount by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Avg Purchase Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# What are the trends in total purchase amount over time (based on Purchase Date)? Are there any seasonal patterns?
```

```python
data['Purchase Date'] = pd.to_datetime(data['Purchase Date'])
```

```python
monthly_sales = data.groupby(data['Purchase Date'].dt.to_period('M'))['Total Purchase Amount'].sum().reset_index()
monthly_sales['Purchase Date'] = monthly_sales['Purchase Date'].dt.to_timestamp()
```

```python
sns.lineplot(data=monthly_sales, x='Purchase Date', y='Total Purchase Amount', marker='o')
plt.title('Monthly Total Purchase Amount Over Time')
plt.xlabel('Month')
plt.ylabel('Total Purchase Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Which product categories generate the highest revenue and which are the most popular based on quantity purchased?
```

```python
popularity_by_category = data.groupby('Product Category')['Quantity'].sum().reset_index()
popularity_by_category = popularity_by_category.sort_values('Quantity', ascending=False)
```

```python
# Revenue
# Calculate total revenue by product category
revenue_by_category = data.groupby('Product Category')['Total Purchase Amount'].sum().reset_index()
revenue_by_category = revenue_by_category.sort_values('Total Purchase Amount', ascending=False)
revenue_by_category = revenue_by_category.rename(columns={'Total Purchase Amount': 'Total Revenue'})

sns.barplot(data=revenue_by_category, x='Total Revenue', y='Product Category')
plt.title('Total Revenue by Product Category')
plt.tight_layout()
plt.show()
```

```python
# What are the most popular payment methods, and how do they relate to the total purchase amount or frequency of purchases?
```

```python
# Total revenue by payment method
revenue_by_method = data.groupby('Payment Method').apply(
    lambda x: (x['Product Price'] * x['Quantity']).sum()
).reset_index(name='Total Revenue')

# Frequency of use
frequency_by_method = data['Payment Method'].value_counts().reset_index()
frequency_by_method.columns = ['Payment Method', 'Transaction Count']
```

```python
sns.countplot(data=data, x='Payment Method', order=data['Payment Method'].value_counts().index)
plt.title('Most Popular Payment Methods')
plt.xlabel('Payment Method')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
revenue_by_method = data.groupby('Payment Method').apply(
    lambda x: (x['Product Price'] * x['Quantity']).sum()
).reset_index(name='Total Revenue')

sns.barplot(data=revenue_by_method, x='Payment Method', y='Total Revenue', order=revenue_by_method.sort_values('Total Revenue', ascending=False)['Payment Method'])
plt.title('Total Revenue by Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Who are the top customers by total purchase amount or frequency, based on Customer ID and Customer Name?
```

```python
top_spenders = data.groupby(['Customer ID', 'Customer Name']).apply(
    lambda x: (x['Product Price'] * x['Quantity']).sum()
).reset_index(name='Total Purchase Amount')
```

```python
sns.barplot(data=top_spenders.head(10), x='Total Purchase Amount', y='Customer Name')
plt.title('Top 10 Customers by Total Spend')
plt.xlabel('Total Purchase Amount')
plt.ylabel('Customer Name')
plt.tight_layout()
plt.show()
```
