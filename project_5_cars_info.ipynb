```python
# Line plot
```

```python
plt.figure(figsize=(8,6))
sns.lineplot(x = 'Price',y = 'Variant',data = data)
plt.title("This is lineplot")
```

```python
# Bar chart or plot
```

```python
sns.barplot(x = 'Price',y = 'Variant',data = data)
plt.title("This is barplot")
plt.show()
```

```python
# scatter plot
```

```python
sns.scatterplot(x = 'Price',y = 'Variant',data = data)
plt.title("This is scatterplot")
plt.show()
```

```python
# Histplot
```

```python
sns.histplot(data['Price'],kde = True)
plt.title("This is histplot")
plt.show()
```

```python
# Boxplot
```

```python
sns.boxplot(data['Price'])
plt.title("This is boxplot")
plt.show()
```

```python
# violinplot
```

```python
sns.violinplot(data['Price'])
plt.title("This is violinplot")
plt.show()
```

```python
# countplot
```

```python
sns.countplot(x = 'Type',data = data)
plt.title("This is countplot")
plt.show()
```

```python
# KDE
```

```python
sns.kdeplot(data['Price'])
plt.title("This is KDE")
plt.show()
```
