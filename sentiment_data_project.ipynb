```python
# Lineplot
```

```python
# Convert Timestamp to datetime object
data['Timestamp'] = pd.to_datetime(data['Timestamp'])

# Group by date and sentiment
daily_counts = data.groupby([data['Timestamp'].dt.date, 'Sentiment']).size().unstack(fill_value=0)

# Plot the lineplot
plt.figure(figsize=(14, 6))
daily_counts.plot(kind='line', marker='o', linewidth=2)
plt.title('Sentiment Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Posts')
plt.legend(title='Sentiment', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Bar chart or plot
```

```python
sentiment_counts = data['Sentiment'].value_counts()

# Plot as bar chart
plt.figure(figsize=(10, 6))
sentiment_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Post Count by Sentiment')
plt.xlabel('Sentiment')
plt.ylabel('Number of Posts')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis='y')
plt.show()
```

```python
# scatterplot
```

```python
sns.scatterplot(x='Country',y='Sentiment',data=data)
plt.title('This is scatterplot')
plt.show()
```
